### Tendência, Sazonalidade e Ruido

Agora é hora de decompor as retas para avaliar seus componentes. 

```python
def decomp_multi_series(parameters: list):
    fig, axes = plt.subplots(4, len(parameters), sharex='col')
    i=0
    plt.rcParams.update({'figure.figsize': (20,12)})
    for column_parameter in parameters:
        decomp = seasonal_decompose(column_parameter[0], model='multiplicative', 
                                   extrapolate_trend='freq')
        axes[0][i].set_title(column_parameter[1])
        decomp.observed.plot(ax=axes[0][i], legend=False, color=column_parameter[2])
        axes[0][i].set_ylabel('Observed')
        decomp.trend.plot(ax=axes[1][i], legend=False, color=column_parameter[2])
        axes[1][i].set_ylabel('Trend')
        decomp.seasonal.plot(ax=axes[2][i], legend=False, color=column_parameter[2])
        axes[2][i].set_ylabel('Seasonal')
        decomp.resid.plot(ax=axes[3][i], legend=False, color=column_parameter[2])
        axes[3][i].set_ylabel('Residual')
        i+=1
    plt.show()
```

```python
parameters = [[co2_in_ar_mod1.data_mean_global, 
               'Global Mole Fration of CO2 Emission in Air', 
               'darkred' ],
              [co2_in_ar_mod1.data_mean_nh, 
               'Northern Mole Fration of CO2 Emission in Air', 
               'darkgreen' ],
              [co2_in_ar_mod1.data_mean_sh, 
               'Southern Mole Fration of CO2 Emission in Air', 
               'darkblue' ]
             ]
decomp_multi_series(parameters)
```

Com a decomposição das retas, fica claro que elas apresentam uma componente sazonal. Logo, para a modelagem destes dados será usado um modelo SARIMA. Ainda assim, é necessário verificar a periodicidade dos dados.

```python
def seazonality_multi_series(parameters:list):
    fig, axes = plt.subplots(len(parameters), sharex=False)
    i=0
    plt.rcParams.update({'figure.figsize': (20,12)})
    for column_parameter in parameters:
        axes[i].set_title(column_parameter[1])
        decomp = seasonal_decompose(column_parameter[0], model='multiplicative', 
                                   extrapolate_trend='freq')
        decomp.seasonal.plot(ax=axes[i], legend=False, color=column_parameter[2])
        axes[i].set_ylabel('Seasonal')
        i+=1
    plt.show()
```

```python
parameters = [[co2_in_ar_mod1.data_mean_global['1980':'1990'], 
               'Global Mole Fration of CO2 Emission in Air', 
               'darkred' ],
              [co2_in_ar_mod1.data_mean_nh['1980':'1990'], 
               'Northern Mole Fration of CO2 Emission in Air', 
               'darkgreen' ],
              [co2_in_ar_mod1.data_mean_sh['1980':'1990'], 
               'Southern Mole Fration of CO2 Emission in Air', 
               'darkblue' ]
             ]
seazonality_multi_series(parameters)
```

Olhando para estes dados, fica claro que nosso modelo apresenta uma sazonalidade anual. Esta informação é importante para a configuração do modelo.

### Estacionaridade

Embora fique claro que os gráfico não são estacionaŕios, é interessante quantificar matemáticamente os mesmo para provar isso.

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title):
    result = adfuller(series, autolag='AIC')
    print(title)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
```

```python
adf_test(co2_in_ar_mod1.data_mean_global, 'Global Mole Fration of CO2 Emission in Air')
adf_test(co2_in_ar_mod1.data_mean_nh, 'Northern Mole Fration of CO2 Emission in Air')
adf_test(co2_in_ar_mod1.data_mean_sh, 'Southern Mole Fration of CO2 Emission in Air')
```

O Teste de Dickey-Fuller aumentado apresenta uma hipótese nula de que a tendência não é estacionária. Se o p-valor < 0.05 a série seria estacionária.

### Autocorrelação

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

def plot_acf_pacf(series, color):
    fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
    plot_acf(series, ax=axes[0], color=color);
    plot_pacf(series, ax=axes[1], color=color);
    
```

```python
plot_acf_pacf(co2_in_ar_mod1.data_mean_global, 'Darkred')
plot_acf_pacf(co2_in_ar_mod1.data_mean_nh, 'Darkgreen')
plot_acf_pacf(co2_in_ar_mod1.data_mean_sh, 'Darkblue')
```

A autocorrelação apresenta a relação de uma váriável com ela mesma. Para uma serie temporal não estacionária, este valor tende a se apresentadar em valor elevado. Já que o valor atual possui relação direta com o anterior. Para a definição dos parâmetros de média móvel e auto regressão do modelo, o ideal é verificar os valores de autocorrelação e correlação parcial da serie transformada em estacionária. Entretanto, é mais interessante para o desenvolvimento de um modelo aplicado em produção que esses modelos sejam definidos automáticamente.

### Criando o modelo

```python
import warnings
import itertools
import statsmodels.api as sm
```

Vamos tentar prever os dados de 20 anos a frente. Sendo assim, treinaremos o modelo ano anterior a 1995

```python
train = co2_in_ar_mod1.data_mean_global['1900':'1994']
test = co2_in_ar_mod1.data_mean_global['1995':]
```

```python
train.tail()
```

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
best_aic = float("inf")
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(train,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = model.fit()
            if results.aic < best_aic:
                best_model = results
                best_param = param
                best_param_seasonal = param_seasonal
                best_aic = results.aic
        except:
            continue
```

```python
print('ARIMA{}x{}12 - AIC:{}'.format(best_param, best_param_seasonal, best_aic))
```

```python
#Avaliando parâmetros
print(best_model.summary().tables[1])
```

Avaliando os parâmetros do nosso modelo, é possível ver que todos eles apresentam alta signficância para o modelo. Tendo um p-valor praticamente igual a zero. Além disso, nossos valores de auto regressão do modelo (AR) apresentam maior significância que os valores de média móvel (MA)

```python
best_model.plot_diagnostics(figsize=(14, 12))
plt.show()
```

```python
pred = best_model.get_prediction(start=pd.to_datetime('1995-01-01'), end=pd.to_datetime('2014-12-01'),dynamic=False)
pred_ci = pred.conf_int()
```

```python
plt.figure(figsize=(12,6))
ax = co2_in_ar_mod1.data_mean_global['1960':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()
plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = test

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```
